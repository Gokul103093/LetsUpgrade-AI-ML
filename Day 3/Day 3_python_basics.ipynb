Question 6:
Write a program to demonstrate all the available data types in Python. Hint: Use type() function.

```
print("Example of string")
a=input()
print(type(a))
print("Example of int")
a=input()
b=int(a)
print(type(b))
print("Example of complex")
b=complex(a)
print(b)
print(type(b))
print("Example of float")
b=float(a)
print(b)
print(type(b))
print("Example of Boolean")
b=bool(input())
print(b)
print(type(b))
print("Example of list")
b=list(input())
print(b)
print(type(b))
print("Example of tuple")
b=tuple(b)
print(b)
print(type(b))
print("Example of set")
b=set(b)
print(b)
print(type(b))
print("Example of dict")
b = {"name" : "gokul", "city" : "chennai"}
print(b)
print(type(b))
```
